```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread("images/pizza_bluescreen.jpg")
```

```python
print(f"This image is {type(image)} , with Dimensions {image.shape}")
```

```python
plt.imshow(image)
```

it's recommeded to take a copy of the image before doing any manpulation on 

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

## Define the color threshold

```python
lower_blue = np.array([0,0,220])
upper_blue = np.array([50, 70, 255])
```

## Create a mask

```python
mask = cv2.inRange(image, lower_blue, upper_blue)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
masked_image = np.copy(image)
```

```python
masked_image[mask !=0] = [0,0,0]
```

```python
plt.imshow(masked_image)
```

## Mask and add a background image

```python
background = cv2.imread("images/space_background.jpg")
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
background = background[0:image.shape[0], 0:image.shape[1]]
```

```python
plt.imshow(background)
```

```python
background[mask == 0 ] = [0, 0, 0]
```

```python
plt.imshow(background)
```

```python
complete_image = background + masked_image
```

```python
plt.imshow(complete_image)
```
