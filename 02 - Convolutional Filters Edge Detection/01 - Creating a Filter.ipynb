```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline

# Read in the image
image = mpimg.imread('images/city_hall.jpg')

plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

### Create Custom Kernel

```python
# sobel filter
kernel = np.array([
[-1, 0, 1],
[-2, 0, 2],
[-1, 0, 1],
])
```

```python
# -1 means that the ouput will have the same type of input image
filtered_image = cv2.filter2D(gray, -1, kernel)
```

```python
plt.imshow(filtered_image, cmap='gray')
```

```python
# create a threshold that sets all filtered pixels to white

retval , binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary_image, cmap='gray')
```
